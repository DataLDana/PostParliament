# load 

```python
# load info
infos_final= pd.read_csv('Ergebnisse/infos_final.csv')
infos_final = infos_final.drop('Unnamed: 0',axis=1)

# load election results
elections = pd.read_csv('Ergebnisse/election2021.csv')
elections = elections.drop('Unnamed: 0',axis=1)
```

```python
# load post chunks
posts =[]
i=0
# read filenames
FileNames = pd.read_csv("Ergebnisse/app/PostFilenames.csv")

for path in FileNames['path']:
    # load the csv
    chunk = change_types(pd.read_csv(path))   # use function to chop datatypes
    # reset index of posts dataframes
    chunk = chunk.set_index('Unnamed: 0')             # get index of original dataframe
    chunk.index.name = None                  # delete the index name
    # append chunk to list of dataframes
    posts.append(chunk)
    i+=1
```

```python
import warnings
show = ['name', 'date', 'likes', 'comments', 'video_views', 'comment', 'webpage'] #,'shortcode'
# write the dataframe according to choosen politicians
df3=pd.DataFrame(columns=show)   # empty df for concatenating

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    # Code that might trigger warnings in pandas
    for i in range(0,len(posts)):    # loop all chunks
        df = posts[i].loc[(mask_pol[i])&(mask_date_max[i])]  # write a df with only masked 
        df3=pd.concat([df3, df],axis=0)      # concatenate from all dfs


timeagg = df3.resample('ME', on='date')['likes', 'comments', 'video_views'].sum().reset_index()



```

```python
show = ['name','party', 'date', 'likes', 'comments', 'video_views'] #,'shortcode'

# write the dataframe according to choosen politicians
df2=pd.DataFrame(columns=show)   # empty df for concatenating

with warnings.catch_warnings():       # supress the warning with pd.concat
    warnings.simplefilter("ignore")
    for i in range(0,len(posts)):    # loop all chunks
        df = posts[i].loc[(posts[i]['party'].isin(parties)) & (mask_date_max[i])] # write a df with only masked (mask_pol[i]) & 
        df2=pd.concat([df2, df],axis=0)      # concatenate from all dfs

# get number of persons in a party
aggregate2 = (
    df2
    .groupby('party')
    .agg(
        post_cnt=('party', 'count'),
        mean_likes=('likes', lambda x: round(x.mean(),0)),
        mean_comments=('comments', lambda x:  round(x.mean(),0))
        )
    .reset_index()
)
aggregate2#['politician']
```
